```julia
using DataFrames
using Gadfly
using Compose
using HypothesisTests
set_default_plot_size(32cm, 10cm)

main_dir = "/home/german/labo/16/ANA/examples/1prn/"
frame_cnt = 600
```

## 1prn

```julia
# MDpocket
mdp_prn_vol = readdlm("mdpocket/mdpout_descriptors.txt", header=true)[1][:, 1:2]
# For some reason MD pocket gives output for 599 frames, instead of 600
mdp_prn_vol = [ mdp_prn_vol; [600.0 mdp_prn_vol[598, 2] ] ]

# POVME
pov_prn_vol = convert(Array{Float64, 2}, readdlm("povme/output/POVME_output.txt")[(end-600):end-1, [1, 3]])

# Epock
epo_prn_vol = readdlm(string(main_dir, "epock/volume.dat"), header = true)[1];

# ANA
loprec_ana_prn_vol = readdlm(string(main_dir, "ana/volumen_loprec"));
hiprec_ana_prn_vol = readdlm(string(main_dir, "ana/volumen_hiprec"));
```

### Correlations

```julia
[ cor(loprec_ana_prn_vol[:,2], mdp_prn_vol[:,2]) cor(hiprec_ana_prn_vol[:,2], mdp_prn_vol[:,2]) cor(mdp_prn_vol[:,2], pov_prn_vol[:,2]) cor(mdp_prn_vol[:,2], epo_prn_vol[:,2]) ]
```

```julia
cor_prn_ana_mdp = cor(loprec_ana_prn_vol[:,2], mdp_prn_vol[:,2])
cor_prn_ana_pov = cor(loprec_ana_prn_vol[:,2], pov_prn_vol[:,2])
cor_prn_ana_epo = cor(loprec_ana_prn_vol[:,2], epo_prn_vol[:,2])

[ cor_prn_ana_epo cor_prn_ana_pov cor_prn_ana_mdp ]
```

```julia
cor_prn_ana_mdp = cor(hiprec_ana_prn_vol[:,2], mdp_prn_vol[:,2])
cor_prn_ana_pov = cor(hiprec_ana_prn_vol[:,2], pov_prn_vol[:,2])
cor_prn_ana_epo = cor(hiprec_ana_prn_vol[:,2], epo_prn_vol[:,2])

[ cor_prn_ana_epo cor_prn_ana_pov cor_prn_ana_mdp ]
```

### Volume

```julia
df_prn_vol = DataFrame(Volume = [ mdp_prn_vol[:,2] ; pov_prn_vol[:,2] ; epo_prn_vol[:, 2];
    loprec_ana_prn_vol[:, 2]; hiprec_ana_prn_vol[:, 2] ],
    Frame = [ collect(1:600); collect(1:600); collect(1:600) ; collect(1:600) ; collect(1:600) ],
    Tool = repeat(["MDP" ; "POVME" ; "Epock"; "ANA:0"; "ANA:1"], inner=[length(hiprec_ana_prn_vol[:, 1])]));


pdf_prn_vol = plot(df_prn_vol, Geom.line, x = "Frame", y = "Volume", color="Tool",
    
    Guide.title("1PRN 600 frames volume calculation"),
    Guide.xlabel("Frame"),
    Guide.ylabel("Volume (Å<\sup>3</sup>)"),
    Coord.cartesian(ymax = 10100),
    Guide.xticks(ticks = collect(100:100:600)),
    Guide.yticks(ticks = collect(2000:4000:10000)),

    Theme(minor_label_font_size = 42pt, major_label_font_size = 48pt,
        key_title_font_size=46pt, key_label_font_size=42pt))
```

### timings

```julia
# MDpocket
suma = 0
for i = 1:3
    proto = readdlm(string("mdpocket/", i,"timing"))[1, 2]
    mins = parse(Float64, proto[1:1])
    secs = parse(Float64, proto[3:end-1])
    suma += frame_cnt / (mins*60 + secs)
end
mdp_prn_rate = suma / 3

# POVME
suma = 0
for i = 1:3
    proto = readdlm(string("povme/", i,"timing"))[1, 2]
    mins = parse(Float64, proto[1:1])
    secs = parse(Float64, proto[3:end-1])
    suma += frame_cnt / (mins*60 + secs)
end
povme_prn_rate = suma / 3

# Epock
suma = 0
for i = 1:3
    proto = readdlm(string("epock/", i,"timing"))[1, 2]
    mins = parse(Float64, proto[1:1])
    secs = parse(Float64, proto[3:end-1])
    suma += frame_cnt / (mins*60 + secs)
end
epock_prn_rate = suma / 3

# ANA lo prec
suma = 0
for i = 1:3
    proto = readdlm(string("ana/", i,"timing_loprec"))[1, 2]
    mins = parse(Float64, proto[1:1])
    secs = parse(Float64, proto[3:end-1])
    suma += frame_cnt / (mins*60 + secs)
end
loana_prn_rate = suma / 3
# ANA hi prec
suma = 0
for i = 1:3
    proto = readdlm(string("ana/", i,"timing_hiprec"))[1, 2]
    mins = parse(Float64, proto[1:1])
    secs = parse(Float64, proto[3:end-1])
    suma += frame_cnt / (mins*60 + secs)
end
hiana_prn_rate = suma / 3

prn_rate = DataFrame(Rate = [ mdp_prn_rate ; povme_prn_rate ; epock_prn_rate ; loana_prn_rate ; hiana_prn_rate ],
Tool = ["MDP"; "POVME"; "Epock" ; "ANA:0" ; "ANA:1"])
```

```julia
pdf_prn_time = plot(prn_rate, Geom.bar, x = "Tool", y = "Rate", color="Tool",

    Guide.title("1PRN 600 frames calculation performance "), 
    
    Coord.Cartesian(xmax=6), 
    Guide.yticks(ticks = [0; 10; 20; 30]),
    Guide.ylabel("Rate (frames / s)"),

    Theme(minor_label_font_size = 42pt, major_label_font_size = 48pt,
        key_title_font_size=46pt, key_label_font_size=42pt))
```
