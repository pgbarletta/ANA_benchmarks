```julia
using DataFrames
using Gadfly
using Compose
set_default_plot_size(32cm, 10cm)
```

## naf_apo

```julia
# POVME
pov_nafapo_vol = readdlm("../naf_apo/povme/output/POVME_output");

# Epock
epo_nafapo_vol = readdlm("../naf_apo/epock/volume.dat");

# MDpocket
mdp_nafapo_vol = readdlm("../naf_apo/mdpocket/2ndrun/mdpout_descriptors.txt", header=true)[1][:, 1:2]
# For some reason MD pocket gives output for 599 frames, instead of 600
mdp_nafapo_vol = [ mdp_nafapo_vol; [600.0 mdp_nafapo_vol[598, 2] ] ]

# ANA
ana_nafapo_vol = readdlm("../naf_apo/ANA_naf_apo_mdvol");
newana_nafapo_vol = readdlm("../naf_apo/newANA_naf_apo_mdvol");

nafapo_vol = DataFrame(Volume = [ pov_nafapo_vol[:, 2]; epo_nafapo_vol[:, 2]; mdp_nafapo_vol[:, 2]; 
    ana_nafapo_vol[:, 2]; newana_nafapo_vol[:, 2] ],
    Software = repeat(["POVME"; "Epock"; "MDP"; "ANA:0"; "ANA:1"], inner=[length(ana_nafapo_vol[:, 1])]));
```

```julia
cor_nafapo_ana_pov = cor(ana_nafapo_vol[:,2], pov_nafapo_vol[:,2])
cor_nafapo_ana_epo = cor(ana_nafapo_vol[:,2], epo_nafapo_vol[:,2])
cor_nafapo_ana_mdp = cor(ana_nafapo_vol[:,2], mdp_nafapo_vol[:,2])
[ cor_nafapo_ana_pov cor_nafapo_ana_epo cor_nafapo_ana_mdp ]
```

```julia
cor_nafapo_newana_pov = cor(newana_nafapo_vol[:,2], pov_nafapo_vol[:,2])
cor_nafapo_newana_epo = cor(newana_nafapo_vol[:,2], epo_nafapo_vol[:,2])
cor_nafapo_newana_mdp = cor(newana_nafapo_vol[:,2], mdp_nafapo_vol[:,2])
[ cor_nafapo_newana_pov cor_nafapo_newana_epo cor_nafapo_newana_mdp ]
```

```julia
pdf_nafapo_vol = plot(nafapo_vol, Geom.line, 
x = [ collect(1:600); collect(1:600); collect(1:600); collect(1:600); collect(1:600)],
    y = "Volume", color="Software",
Guide.xlabel("Frame"), Guide.title("4UET 600 frames volume calculation"),
Theme(minor_label_font_size = 14pt, major_label_font_size = 14pt,
key_title_font_size=12pt, key_label_font_size=12pt))
```

### timings

```julia
# POVME
pov_nafapo_time = readdlm("../naf_apo/povme/parallel_tiempo");
# Epock
epo_nafapo_time = readdlm("../naf_apo/epock/tiempo");
# MDpocket
mdp_nafapo_time = readdlm("../naf_apo/mdpocket/tiempo")
# ANA
ana_nafapo_time = readdlm("../naf_apo/tiempo");
nafapo_time = DataFrame(Duration = [ pov_nafapo_time[1,1]; epo_nafapo_time[1,1]; mdp_nafapo_time[1,1]; ana_nafapo_time[1,1] ],
Software = ["POVME"; "Epock"; "MDP"; "ANA"])
```

```julia
pdf_nafapo_time = plot(nafapo_time, Geom.bar, x = "Software", y = "Duration", color="Software",
Guide.title("4UET 600 frames calculation performance "), Guide.ylabel("Duration (s)"),
Coord.Cartesian(ymax=900, xmax=5), Guide.yticks(ticks = [0; 250; 500; 750]),
Guide.annotation(compose(compose(context(), text(0.7, 750, "4 processors\n   12mins")), fontsize(28pt), fill("black"))),
Guide.annotation(compose(compose(context(), text(1.8, 750, "8.4secs")), fontsize(34pt), fill("black"))),
Guide.annotation(compose(compose(context(), text(2.8, 750, "9.5mins")), fontsize(28pt), fill("black"))),
Guide.annotation(compose(compose(context(), text(3.8, 750, "14.5secs")), fontsize(28pt), fill("black"))),
Theme(minor_label_font_size = 32pt, major_label_font_size = 24pt, key_position=:right,
key_title_font_size=26pt, key_label_font_size=32pt))
```

### save PDFs and PNGs

```julia
out_filename_pdf_nafapo_vol = "4uet_vol.pdf"
out_filename_pdf_nafapo_time = "4uet_time.pdf"
draw(PDF(out_filename_pdf_nafapo_vol, 40cm, 20cm), pdf_nafapo_vol)
draw(PDF(out_filename_pdf_nafapo_time, 40cm, 20cm), pdf_nafapo_time)
```

```julia
out_filename_pdf_nafapo_vol = "4uet_vol.png"
out_filename_pdf_nafapo_time = "4uet_time.png"
draw(PNG(out_filename_pdf_nafapo_vol, 1800px, 800px), pdf_nafapo_vol)
draw(PNG(out_filename_pdf_nafapo_time, 1800px, 800px), pdf_nafapo_time)
```
